```python
from Bio import SeqIO

with open("sample.txt",'r') as file:
    sample=file.read()

def get_seq(id,st,len):
    file=SeqIO.parse("./reference/reference.fasta","fasta")
    i=0
    for seq_record in file:
        if i==id:
            print(seq_record.id)
            return seq_record.seq[st:st+len]
        i+=1

# 5,160600000
s=get_seq(5,160600000,len(sample)*2)
with open("reference.txt",'w') as file:
    file.write(str(s))
```
